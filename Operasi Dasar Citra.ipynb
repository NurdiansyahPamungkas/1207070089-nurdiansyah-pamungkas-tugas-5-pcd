```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert

import numpy as np
```

Percobaan 1 - Cropping Image

```python
gambar1 = plt.imread('184637_1.jpg')
gambar2 = plt.imread('0000467563.jpg')

gambar1Cropped = gambar1.copy()
gambar1Cropped = gambar1Cropped[0:256,64:320]

gambar2Cropped = gambar2.copy()
gambar2Cropped = gambar2Cropped[64:256,128:320]

print('gambar1 Ori Shape : ',gambar1.shape)
print('gambar1 Crop Shape : ',gambar1Cropped.shape)

print('gambar2 Ori Shape : ',gambar2.shape)
print('gambar2 Crop Shape : ',gambar2Cropped.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(gambar1)
ax[0].set_title("Citra Input 1")

ax[1].imshow(gambar2, cmap='gray')
ax[1].set_title('Citra Input 2')

ax[2].imshow(gambar1Cropped)
ax[2].set_title("Citra Output 1")

ax[3].imshow(gambar2Cropped, cmap='gray')
ax[3].set_title('Citra Output 2')
```

Percobaan 2 - Citra Negative

```python
inv = invert(gambar1Cropped)
print('Shape Input : ', gambar1Cropped.shape)
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(gambar1Cropped)
ax[0].set_title("Citra Input")

ax[1].hist(gambar1Cropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
copygambar2 = gambar2Cropped.copy().astype(float)

shape = copygambar2.shape

output1 = np.empty(shape)

for baris in range(0, shape[0]-1):
    for kolom in range(0, shape[1]-1):
        a1 = baris
        b1 = kolom
        output1[a1, b1] = copygambar2[baris, kolom] / 100
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(gambar2Cropped, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].hist(gambar2Cropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(output1, cmap='gray')
ax[2].set_title('Citra Output (Brightness)')

ax[3].hist(output1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
